```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
plt.rcParams['font.family'] = 'Times New Roman'

# 
morandi_colors = [
                  '#3CA4CF', '#399FD1', '#379BD3', '#3596D5', '#3391D7'
                  , '#318CD9', '#2F87DA', '#2D82D8', '#2B7DD6', '#2978D4'
                  , '#F1C8B6', '#EAAEA0', '#E1938A', '#D97977', '#D16067', '#C84758'
                  ,'#B3D3C6', '#8AB4A0', '#639680', '#3D7760'
                  
]
# 
sample_data = {
    'Melting temperature': 2079,
    'Glass transition temperature': 5072,
    'Decomposition temperature': 3520,
    'Thermal conductivity': 78,
    'Heat capacity': 79,
    'Atomization energy': 390,
    'Limiting oxygen index': 101,
    'Crystallization tendency': 543,
    'Density': 910,
    'Fractional free volume': 128,
    'Bandgap': 3941,
    'Electron affinity': 368,
    'Ionization energy': 370,
    'Refractive index': 898,
    'Dielectric constant': 382,
    'Frequency dependent electric constant': 1187,
    'Tensile strength': 672,
    'Youngs modulus': 629,
    'Hildebrand solubility parameter': 112,
    'Gas permeability': 2168
}

# 
class_names = []
sample_counts = []
child_sample_counts = []
for class_name, samples in sample_data.items():
    class_names.append(class_name)
    if isinstance(samples, list):
        sample_counts.append(sum(samples))
        child_sample_counts.append(len(samples))
    else:
        sample_counts.append(samples)
        child_sample_counts.append(0)

# 
fig, ax = plt.subplots(figsize=(20, 19))  # 

# 
ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)


# 
font = FontProperties(family='Times New Roman', style='normal', size=35)

# 
y_pos = np.arange(len(class_names))
bars = ax.barh(y_pos, sample_counts, align='center')

# 
for i, bar in enumerate(bars):
    bar.set_color(morandi_colors[i % len(morandi_colors)])

# 
# ax.set_yticks(y_pos)
# ax.set_yticklabels(class_names, fontproperties=font, fontsize=20)  # 
ax.set_xlabel('Count', fontproperties=font)  # 

# 
ax.tick_params(axis='x', labelsize=30, which='major')
# ax.tick_params(axis='y', labelsize=15, which='major')

# 
ax.legend(reversed(bars), reversed(class_names), fontsize=30
          , ncol=2, loc='upper right',framealpha=1.0
          ,columnspacing=0.4, handletextpad=0.1
          , bbox_to_anchor=(1, 0.92))  # 

plt.yticks([])
# 
plt.savefig('./polymer_datasets2.jpg', dpi=1500)
plt.show()
```
